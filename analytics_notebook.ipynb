# Построим таблицу указанную в ТЗ

```python
list_content = []

for i, cell in enumerate(date['cells']):
    cell_id = i
    type_cell = cell['cell_type']
    content_cell = cell['source']

    list_content.append([cell_id, type_cell, content_cell])
    
df_content_notebook = pd.DataFrame(list_content, 
                                   columns=['Id', 
                                            'Type', 
                                            'Content'])
```

```python
df_content_notebook
```

# Определим язык текста в ячейках с markdown

```python
from langdetect import detect_langs
import re
```

```python
count_value = len(df_content_notebook)

for index in range(count_value):
    line = dict(df_content_notebook.loc[index]) 
    if line['Type'] == 'markdown':
        id_line = line['Id']
        for text in line['Content']:
            # Проверим, состоит ли строка из текста
            if bool(re.match('^(?=.*[a-zA-Z])', text)):
                # Проверим, состоит ли строка только из ссылки
                if re.search("(?P<url>https?://[^\s]+)", text) is None:
                    print(f'{id_line} -- {detect_langs(text)}')
```

# Определим количество вызовов функций в ячейке

Допущения:
1. Будем считать, что нет многострочных комментариев
2. Будем считать, что вызов функции и скобки этой функции указаны на одной строке

Алгоритм: 
1. Удалим комментарии
2. Будем обходить строки и смотреть, есть ли в ней конструкции вида func()

```python
def remove_comments(line,sep="#"):
    i = line.find(sep)
    if i == 0 :
        line = ''
    elif i > 0:
        line = line[:i]
    return line
```

```python
ount_value = len(df_content_notebook)

count_function_calls = 0

for index in range(count_value):
    line = dict(df_content_notebook.loc[index]) 
    if line['Type'] == 'code':
        for text in line['Content']:
            text_without_comment = remove_comments(text)
            r = re.findall('\w*\([\w\,\s]*\)', text_without_comment)
            count_function_calls += len(r)
            
print(f'Количество вызовов функций: {count_function_calls}')
```
